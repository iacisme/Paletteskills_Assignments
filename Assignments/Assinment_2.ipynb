```python
import random
import pandas as pd
# Library to find and identify missing values
import missingno as msno 
# Library to create interactive tools
import ipywidgets as widgets
# Plotly graphing library
import plotly.express as px
import plotly.graph_objs as go

import warnings
warnings.filterwarnings('ignore')

# Define the range for longitude and latitude
lon_range = (-105, -100)
lat_range = (50, 52)

# Define the list of soil types
soil_types = ['black', 'gray', 'dark brown', 'brown', None]

# Create an empty list to store the farms
farms = []

# Generate 40 farms
for i in range(1000):
    farm = {
        'farm_name': f'Farm {i+1}', # f is formated string
        'farm_lon': round(random.uniform(lon_range[0], lon_range[1]), 2), # random lon with 2 decimal points
        'farm_lat': round(random.uniform(lat_range[0], lat_range[1]), 2), # rantom lat with 2 decimal points
        'temperature': round(random.uniform(18, 33), 1), #random temp with 1 decimal points
        'precipitation': round(random.uniform(20, 90), 1), # random prec with 1 decimal points
        'soil_type': random.choice(soil_types) # random soil types from above list
    }
    farms.append(farm)
```

 4. Calculate the standard deviation of precipitation for each soil type.

```python
## Your code here down

def calculateSTDbySoilColour(soilTypes_list):
    """
        Function works by taking a list of soil types and then looping through the list 
        to calculate the standard deviation of each soil type
    """

    # Calculate the standard deviation by soil type
    for soilType in soilTypes_list:
        # Get the farms by soil type
        soilTypesByFarm_df = farms_df[farms_df['soil_type'] == soilType]
        # Calculate the standard deviation of the precipitation of the soil type
        precipitation_standardDeviation = soilTypesByFarm_df['precipitation'].std()
        # Display the results
        message = f"The standard deviation of the precipitation of {soilType} soil colour is: {precipitation_standardDeviation:.2f}"
        print(message)
        print()

# Create a list of soil types
soilTypes_list = farms_df['soil_type'].unique().tolist()

# Calculate the standard deviation of precipitation by soil colour
calculateSTDbySoilColour(soilTypes_list)

## Your code here up
```

5. Calculate the cumulative sum of precipitation for each soil type.

15. Plot the distribution of relevant columns.

```python
## Your code here down

# Creat a list of feature names
features_list = updatedFarms_df.keys().to_list()
# Use the list to create a Dropdown widget
features_dropDown = widgets.Dropdown(
    options = features_list,
    value = features_list[0],
    description = 'Feature:',
    disabled = False,
)

# Create a list of labels
labels_list = features_list.copy()
# Insert None as the first value of this list
labels_list.insert(0, None)
# Use the list to create a Dropdown widget
labels_dropDown = widgets.Dropdown(
    options = labels_list,
    value = labels_list[0],
    description = 'Label:',
    disabled = False,
)

# List of mariginals to use with the histogram
marginals_list = [
    None,
    'box',
    'violin',
    'rug',
]
# Use the list to create RadioButtons
marginal_radioBtns = widgets.RadioButtons(
    options = marginals_list,
    value = marginals_list[0], 
#    layout={'width': 'max-content'}, # If the items' names are long
    description = 'Marginals:',
    disabled = False
)

# Create a Histogram Interactive Plot
@widgets.interact(
    feature = features_dropDown,
    label = labels_dropDown,
    marginals = marginal_radioBtns 
)
def histogramFunction(feature, label, marginals):
    fig = px.histogram(
        updatedFarms_df,
        x = feature,
        color = label,
        marginal = marginals 
    )

    fig.update_layout(
        title_text = f"Histogram of the distribution of feature: {feature}",
        width = 800,
        height = 600
    )

    fig.show()
    
## Your code here up
```
